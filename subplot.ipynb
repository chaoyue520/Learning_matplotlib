# subplot

* [参考](https://segmentfault.com/a/1190000005104723)
* 本质是，先通过参数 $axes$ 定义好空白画图，然后在使用plot画图的时候，将具体图像plot到指定的画布中，该过程通过参数 $ax$ 控制

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 搞点测试数据
df1=pd.DataFrame(numpy.random.randn(10))
df2=pd.DataFrame(numpy.random.randn(10))
df3=pd.DataFrame(numpy.random.randn(10))
df4=pd.DataFrame(numpy.random.randn(10))
```

## 常规子图：均匀画图

### 方法1：

```python
# 先设定2*2的画布
fig,axes = plt.subplots(2,2)

df1.plot(kind='box',ax=axes[0,0],figsize=(5,5))
df2.plot(kind='density',ax=axes[0,1],figsize=(5,5))
df3.plot(kind='bar',ax=axes[1,0],figsize=(5,5))
df4.plot(kind='line',ax=axes[1,1],figsize=(5,5))
```

### 方法2：

```python
# 先指定每个ax在画布上所在的位置，然后再填充内容
ax1=plt.subplot(2,2,1) # 将画布分成2行2列，图形放置在第1个位置
ax2=plt.subplot(2,2,2) # 将画布分成2行2列，图形放置在第2个位置
ax3=plt.subplot(2,2,3) # 将画布分成2行2列，图形放置在第3个位置
ax4=plt.subplot(2,2,4) # 将画布分成2行2列，图形放置在第4个位置

df1.plot(kind='box',ax=ax1,figsize=(5,5))
df2.plot(kind='density',ax=ax2,figsize=(5,5))
df3.plot(kind='bar',ax=ax3,figsize=(5,5))
df4.plot(kind='line',ax=ax4,figsize=(5,5))
```

### 方法3

```python
# 两行一列第一个位置
ax1=plt.subplot(211)
# 两行一列第二个位置
ax2=plt.subplot(212)

df1.plot(kind='box',ax=ax1)
df2.plot(kind='bar',ax=ax2)
```

## 更加灵活：不均匀画图

### 第一类

```python
ax1=plt.subplot(2,2,1) # 将画布分成2行2列，图形放置在第1个位置
ax2=plt.subplot(2,2,2) # 将画布分成2行2列，图形放置在第2个位置
ax3=plt.subplot(2,1,2) # 将画布分成2行1列，图形放置在第2个位置

df1.plot(kind='kde',ax=ax1)
df2.plot(kind='box',ax=ax2)
df3.plot(kind='density',ax=ax3)

# plt.show()
```

### 第二类

```python
ax1=plt.subplot(2,1,1) # 将画布分成2行1列，图形放置在第1个位置，本质上已经占了两个位置，所以ax2和ax3只能从3,4开始排列
ax2=plt.subplot(2,2,3) # 将画布分成2行3列，图形放置在第2个位置
ax3=plt.subplot(2,2,4) # 将画布分成2行3列，图形放置在第3个位置


df1.plot(kind='kde',ax=ax1)
df2.plot(kind='bar',ax=ax2)
df3.plot(kind='box',ax=ax3)
```

### 第三类

```python
ax1=plt.subplot(2,1,1) # 将画布分成2行1列，图形放置在第1个位置，本质上已经占了三个位置，所以ax2和ax3只能从4开始排列
# 第2行的图只能从4开始
ax2=plt.subplot(2,3,4) # 将画布分成2行3列，图形放置在第2个位置
ax3=plt.subplot(2,3,5) # 将画布分成2行3列，图形放置在第3个位置
ax4=plt.subplot(2,3,6) # 将画布分成2行3列，图形放置在第3个位置

df1.plot(kind='density',ax=ax1)
df2.plot(kind='box',ax=ax2)
df3.plot(kind='kde',ax=ax3)
df4.plot(kind='bar',ax=ax4)
```

### 第四类

```python
ax1=plt.subplot(2,3,1) # 将画布分成2行3列，图形放置在第1个位置
ax2=plt.subplot(2,3,2) # 将画布分成2行3列，图形放置在第2个位置
ax3=plt.subplot(2,3,3) # 将画布分成2行3列，图形放置在第3个位置

ax4=plt.subplot(2,2,3) # 将画布分成2行1列，图形放置在第1个位置
ax5=plt.subplot(2,2,4) # 将画布分成2行1列，图形放置在第2个位置


df1.plot(kind='kde',ax=ax1)
df2.plot(kind='box',ax=ax2)
df3.plot(kind='density',ax=ax3)
df4.plot(kind='bar',ax=ax4)
df4.plot(kind='barh',ax=ax5)

# plt.show()
```
